```python
import pandas as pd
import pymongo
import datetime
```

```python
client = pymongo.MongoClient()
database = client['evaluation']
collection = database['edf']
```

```python
df_matches = pd.read_csv("../data/df_matches.csv")
df_matches.head(20)
```

# Nettoyage des datas

```python
df_matches['date'] = pd.to_datetime(df_matches['date'], errors='coerce')
df_matches = df_matches.dropna(subset=['date'])

df_matches['_id'] = df_matches.index
df_matches.rename(columns={'X4':'affiche','X5':'score_final','X6':'match_type'}, inplace=True)
l_d =df_matches.to_dict(orient='records')


 
#df_matches = df_matches.drop_duplicates(subset='affiche', keep='last')
```

# Import des datas

```python
from pymongo import MongoClient

client = MongoClient("mongo:27017")

db_mt = client.mt

collection_mt = db_mt['matches']

collection_mt.insert_many(l_d)
```

# Requête 0

```python
req = collection_mt.find()
for test in req:
    print(test)
```

# Requête 1

```python
req = collection_mt.find({"outcome":"loss"}).limit(5)
next(req)
```

# Requête 2

```python
req = collection_mt.find({"outcome":"win"})
req.count()
```

# Requête 3

```python
req = collection_mt.aggregate([{"$group" : {"_id" : "$win", "Penaltys" : {"$sum" : 1}}}])
list(req)
```

# Requête 4

```python
date = datetime.datetime.strptime("1950-01-01",'%Y-%m-%d')
req = collection_mt.find({"$and":[{'outcome':'loss'}, {'date':{'$lt':date}}]})
list(req)
```

# Requête 5

```python
req = collection_mt.find({"outcome":"win", "match_type":"Match amical"})
req.count()
```

# Requête 7

```python
req = collection_mt.aggregate([{"$group": {"_id" : "$match_type", "matches" : {"$sum" : 1}}}])
list(req)
```

# Requête 8

```python
cur = collection_mt.find({"$and":[{"year":{"$gte": 2000, "$lte":2021}}, {"adversaire":"Belgique"}]})
cur.count()
```
